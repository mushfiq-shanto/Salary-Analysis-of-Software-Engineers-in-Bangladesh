```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/MasterCourse/Capstone #1/eng_salaries.csv')
df.head()
```

```python
df['monthly?'] = df['Frequencies'].apply(lambda x: '/ mo' in x)
df_e = df[df['monthly?'] == False]
display(df_e)
```

```python
def convert_to_float(value):
    if value is None or isinstance(value, float):
        return value

    value = value.replace('BDT', '').replace(',', '').strip()

    if 'K' in value:
        return float(value.replace('K', '')) * 1000
    if 'M' in value:
        return float(value.replace('M', '')) * 1_000_000

    try:
        return float(value)
    except ValueError:
        return value  # Return original value if conversion fails

for column in ['Medians', 'Uppers', 'Lowers']:
    df[column] = df[column].apply(convert_to_float)

df.head()
```

```python
df.loc[df['Frequencies'].str.contains(' / yr'),['Medians','Uppers','Lowers']] /= 12
df['Frequencies'] = df['Frequencies'].str.replace(' / yr', ' / mo', regex=False)
df.head(100)
```

```python
df['Entries'] = df['Salaries'].str.split(' ').str[1].astype(int)
df = df.drop(columns=['Salaries','monthly?'])
df.head()
```

```python
# Rearrange columns to bring 'Entries' as the 2nd column
df = df[['Company', 'Entries', 'Rating', 'Medians', 'Uppers', 'Lowers', 'Frequencies']]

# Rename 'Uppers' and 'Lowers' to 'Upper Range' and 'Lower Range'
df = df.rename(columns={'Uppers': 'Upper Range', 'Lowers': 'Lower Range','Medians':'Median Salary', 'Frequencies':'Salary Interval'})

# Display the updated DataFrame
df.head()
```

```python
df.to_csv('/content/drive/MyDrive/MasterCourse/Capstone #1/Software Engineer Salaries in Bangladesh.csv',index=False)
```
